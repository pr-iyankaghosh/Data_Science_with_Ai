```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'C:\FBS\Data Analysis\Dataexploration\Position_Salaries.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Level'],df['Salary'])
plt.title("Level vs Salary")
plt.xlabel("Level")
plt.ylabel("Salary")
```

```python
X=df["Level"]
y=df["Salary"]
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
regressor.fit(X.values.reshape(-1,1),y)
```

```python
y_pred=regressor.predict(X.values.reshape(-1,1))
```

```python
plt.plot(X.values.reshape(-1,1),y_pred,color='red')
plt.scatter(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
poly_reg.fit(X.values.reshape(-1,1))
x_poly=poly_reg.transform(X.values.reshape(-1,1))
```

```python
regressor2=LinearRegression()
regressor2.fit(x_poly,y)
```

```python
y_pred2=regressor2.predict(x_poly)
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred2,color='red')
```
