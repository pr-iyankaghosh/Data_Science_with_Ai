```python
plt.scatter(df['YearsExperience'],df['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
```

```python
X=df['YearsExperience']#input variable
y=df['Salary']#output variable
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
# model training linearRegression model
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train.values.reshape(-1,1),y_train)
```

```python
# models find the best value of intercept and slope (m andc)
```

```python
m=lr.intercept_
c=lr.coef_
print(m,c)
```

```python
y_pred_train=lr.predict(x_train.values.reshape(-1,1))
```

```python
plt.scatter(x_train,y_train,color='red')
#line plot (linear regressor)
plt.plot(x_train,y_pred_train,color='blue')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
```

```python
# finding errors
from sklearn.metrics import mean_squared_error,root_mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
```

```python
print("Mean Squared Error",mean_squared_error(y_train,y_pred_train))
print("Root Mean Squared Error",root_mean_squared_error(y_train,y_pred_train))
print("Mean Absolute Error",mean_absolute_error(y_train,y_pred_train))
print("Mean Absolute percentage Error",mean_absolute_percentage_error(y_train,y_pred_train))
```
