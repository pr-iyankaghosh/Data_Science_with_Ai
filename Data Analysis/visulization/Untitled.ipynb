```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[10,20,90,56]
plt.title("id and marks")
plt.xlabel("id")
plt.ylabel("marks")
plt.plot(x,y)
plt.show()
```

```python
#scatter plot find the correlation between two values
#it can be name and marks also
plt.scatter(x,y,color='Red',marker="*")
plt.title("id and marks")
plt.xlabel("id")
plt.ylabel("marks")
plt.plot(x,y,color='yellow')
```

```python
x=[1,2,3,4]
y=[10,20,90,56]
plt.title("id and marks")
plt.xlabel("id")
plt.ylabel("marks")
plt.bar(x,y,color='yellow') #helpful to compare two categories
```

```python
plt.barh(x,y,color='yellow')
```

```python
plt.hist(y)
```

```python
#finding contribution percentage wise
plt.pie(y,labels=x)
```
